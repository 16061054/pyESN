```python
import numpy as np
from pyESN import ESN

data = np.load('mackey_glass_t17.npy') 
esn = ESN(n_inputunits = 1,
          n_outunits = 1,
          n_reservoir = 500,
          spectral_radius = 1.5,
          random_state=42)

trainlen = 2000
future = 2000
pred_training = esn.fit(np.ones(trainlen),data[:trainlen])
prediction = esn.predict(np.ones(future))
print("test error: \n"+str(np.sqrt(np.mean((prediction.flatten() - data[trainlen:trainlen+future])**2))))
```

```python
figure(figsize=(11,1.5))
plot(range(0,trainlen+future),data[0:trainlen+future],'k',label="target system")
plot(range(trainlen,trainlen+future),prediction,'r', label="free running ESN")
lo,hi = ylim()
plot([trainlen,trainlen],[lo+np.spacing(1),hi-np.spacing(1)],'k:')
legend(loc=(0.61,1.1),fontsize='x-small')
```
