```python
import numpy as np
from pyESN import ESN

data = np.loadtxt('MackeyGlass_t17.txt')

esn = ESN(n_inputunits = 1,
          n_outunits = 1,
          n_reservoir = 500,
          spectral_radius = 1.5,
          random_state=42,
          silent=False)


trainlen = 2000
pred_training = esn.fit(np.ones(trainlen),data[:trainlen])

future = 2000
prediction = esn.predict(np.ones(future))
print("test error: \n"+str(np.sqrt(np.mean((prediction - data[0:trainlen+future])**2))))
```

```python
figure(figsize=(12,1.5))
plot(range(0,trainlen+future),data[0:trainlen+future],'b',label="target signal")
plot(range(trainlen,trainlen+future),prediction,'r', label="free running ESN")
lo,hi = ylim()
plot([trainlen,trainlen],[lo+np.spacing(1),hi-np.spacing(1)],'k')
legend(loc=(0.875,1.1),fontsize='x-small')
```
